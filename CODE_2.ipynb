```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'ecommerce_stores_50.csv'
df = pd.read_csv(file_path)

# Plot a bar chart showing the number of employees for each store
plt.figure(figsize=(12, 8))
plt.bar(df['Store Name'], df['Employees Count'])
plt.xlabel('Store Name')
plt.ylabel('Employees Count')
plt.title('Number of Employees in Each Store')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'ecommerce_stores_50.csv'
df = pd.read_csv(file_path)

# Sort the dataframe by store name to make the line chart clearer
df = df.sort_values('Store Name')

# Plot a line chart showing the number of employees for each store
plt.figure(figsize=(12, 8))
plt.plot(df['Store Name'], df['Employees Count'], marker='o')
plt.xlabel('Store Name')
plt.ylabel('Employees Count')
plt.title('Number of Employees in Each Store')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
